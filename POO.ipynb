```python
#Asociar un comportamiento
from random import randint
class Tragamonedas:

  def __init__(self, id, premio):
    self.id = id
    self.premio = premio
    self.monedas = 0 #Definimos un atributo
    self.jackpots = 0 #Cuantas veces a dado premio


  def jugar(self):
    self.monedas += 1
    numeros =randint(0,9), randint(0,9)
    mensaje = ""
    if numeros[0] == numeros[1]:
      self.jackpots += 1
      mensaje = "Felicidades !! Ganaste %0.2f" % self.premio
    else:
      mensaje = "No ganaste"
    return numeros, mensaje
```

```python
tragam3 = Tragamonedas(3, 1000)
tragam4 = Tragamonedas(4, 5000)
```

```python
print(tragam3.jugar())
print(tragam4.jugar())
print(tragam3.id, tragam3.jackpots, tragam3.monedas)
print(tragam4.id, tragam4.jackpots, tragam4.monedas)
```

```python
# Agregamos dos acciones, atacar y subir de nivel se generan al azar lanzando un dado
class Personaje:

    def __init__(self, nombre, fuerza, energia, defensa, vida):
        self.nombre = nombre
        self.fuerza = fuerza
        self.energia = energia
        self.defensa = defensa
        self.vida = vida

    def atributos(self):
        print(self.nombre)
        print("-Fuerza ", self.fuerza)
        print("-Energía ", self.energia)
        print("-Defensa ", self.defensa)
        print("-Vida ", self.vida)

    def subir_nivel(self, fuerza, defensa, vida):#Aumenta los atributos
        self.fuerza = self.fuerza + fuerza
        self.defensa = self.defensa + defensa
        self.vida = self.vida + vida


    def ataque(self, enemigo):# Ataca al enemigo
        daño = random.randint(-1, 3)*(self.fuerza - enemigo.defensa)
        self.energia = self.energia - 2
        if daño > 0:
            enemigo.vida = enemigo.vida - daño
            enemigo.defensa = enemigo.defensa - 1
            print(self.nombre, "hizo un daño = ", daño)
        else:
            self.vida = self.vida + daño #Se resta vida propia pues daño<0
            enemigo.energia = enemigo.energia - 2
            self.defensa = self.defensa - 1
            print(enemigo.nombre, "hizo un daño = ", -daño)
```

```python
mi_personaje = Personaje("Carlos", 12, 8, 10, 20)
guerrero = Personaje("Guerrero", 14, 8, 8, 20)
mi_personaje.atributos()
guerrero.atributos()
```

```python
dado = random.randint(1, 3)
print("Dado = ", dado)
if dado == 1:
    mi_personaje.subir_nivel(2, 2, 2)
    print("has subido de nivel")
elif dado == 2:
    print("El guerrero te ha atacado")
    guerrero.ataque(mi_personaje)
else:
    print("Has atacado al guerrero")
    mi_personaje.ataque(guerrero)

mi_personaje.atributos()
guerrero.atributos()

if mi_personaje.vida <= 0:
    print("Has muerto")
if guerrero.vida <= 0:
    print("!Has derrotado al guerrero¡")
if mi_personaje.energia <= 0:
    print("Te has rendido")
if guerrero.energia <= 0:
    print("El guerrero se ha rendido")
```
