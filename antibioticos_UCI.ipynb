# Resultado de los datos de los antibióticos y sus respuestas (UCI)
Se grafican los resultados de las respuestas de los antibióticos en 36 meses, desde julio del 2018 hasta junio del 2021

## Se cargan y se limpian los datos

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
datos = pd.read_excel("datos_antibioticos_UCI.xlsx", index_col=0)
datos = datos.T
datos.head()
```

```python
x = np.linspace(1,36, 36)
```

## Gráfica general de las tablas 1 y 2 (UCI)

```python
plt.figure(figsize=(14,5))
plt.plot(x, datos["VANCOMICINA"], '-', label="VANCOMICINA")
plt.plot(x, datos["CEFTRIAXONA"], '--', label="CEFTRIAXONA")
plt.plot(x, datos["CEFEPIMA"], ':', label="CEFEPIMA")
plt.plot(x, datos["MEROPENEM"], 'v-', label="MEROPENEM")
plt.plot(x, datos["ERTAPENEM"], 'o-', label="ERTAPENEM")
plt.plot(x, datos["PIPERACILINA TAZOBACTAM"], '2-', label="PIPERACILINA TAZOBACTAM")
plt.plot(x, datos["COLISTINA"], '*-', label="COLISTINA")
plt.plot(x, datos["LINEZOLID"], 'h-', label="LINEZOLID")
plt.plot(x, datos["POLIMIXINA B"], 'D-', label="POLIMIXINA B")
plt.xlabel('Tiempo (meses)')
plt.ylabel('Dosis')
plt.title('Tabla-1')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14,5))
plt.plot(x, datos["k pneumonie R meropenem"], '-', label="k pneumonie R meropenem")
plt.plot(x, datos["k pneumonie R píptazo"], '--', label="k pneumonie R píptazo")
plt.plot(x, datos["p.aeruginosa R meropenem"], ':', label="p.aeruginosa R meropenem")
plt.plot(x, datos["p.aeruginosa R piptazo"], 'v-', label="p.aeruginosa R piptazo")
plt.plot(x, datos["sarrm"], 'o-', label="sarrm")
plt.xlabel('Tiempo (meses)')
plt.ylabel('Resistencia')
plt.title('Tabla-2')
plt.legend()
plt.show()
```

## Correlación

La correlación lineal de Pearson nos da una idea de que par de variables se relacionan. Si el valor de la correlación es cercano a 1 o a -1, entonces la correlación es fuerte.

```python
cor_1 = pd.read_excel("UCI_correlacion_1.xlsx", index_col=0)
cor_1 = cor_1.T
```

```python
plt.figure(figsize=(6,6))
matriz = cor_1.corr()
plt.matshow(matriz, cmap="bwr", vmin=-1, vmax=1, fignum=1)
plt.xticks(range(3), cor_1.columns, rotation=90)
plt.yticks(range(3), cor_1.columns)
for i in range(len(cor_1.columns)):
    for j in range(len(cor_1.columns)):
        plt.text(i, j, round(matriz.iloc[i, j],2))
plt.colorbar()
plt.show()
```
