```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
sns.get_dataset_names();
```

```python
pinguinos = sns.load_dataset("penguins")
pinguinos.columns = ["especie", "isla", "longitud_pico",
                     "profundidad_pico", "longitud_aleta", "masa", "sexo"]
```

```python
pinguinos.count()
pinguinos.head()
pinguinos.describe()
```

```python
sns.relplot(data=pinguinos)
plt.show()
```

```python
sns.relplot(data=pinguinos,
           x="longitud_pico",
           y="longitud_aleta", height=4)
plt.show()
```

```python
sns.relplot(data=pinguinos,
           x="longitud_pico",
           y="longitud_aleta",
           hue="masa")
plt.show()
```

```python
sns.relplot(data=pinguinos,
           x="longitud_pico",
           y="longitud_aleta",
           hue="especie")
plt.show()
```

```python
sns.relplot(data=pinguinos,
           x="longitud_pico",
           y="longitud_aleta",
           hue="especie",
            size="masa",
            sizes=(100, 500),
           style="sexo")
plt.show()
```

```python
sns.relplot(data=pinguinos,
           x="longitud_pico",
           y="longitud_aleta",
           hue="especie",
            row="sexo",
            size="masa",
            sizes=(100, 500),
           style="sexo")
plt.show()
```

```python
sns.violinplot(data=pinguinos, x="longitud_aleta", y="isla")
plt.show()
```

```python
sns.catplot(data=pinguinos, x="isla", y="longitud_aleta", kind='strip', hue='sexo')
plt.show()
```

```python
pinguinos.drop("especie", axis=1, inplace=True)
pinguinos.drop("isla", axis=1, inplace=True)
pinguinos.drop("sexo", axis=1, inplace=True)
```

```python
pinguinos.corr()
```

```python
matriz = pinguinos.corr()
plt.matshow(matriz, cmap="bwr", vmin=-1, vmax=1)
plt.xticks(range(4), pinguinos.columns, rotation=90)
plt.yticks(range(4), pinguinos.columns)
plt.colorbar()
plt.show()
```

```python
matriz = pinguinos.corr()
plt.matshow(matriz, cmap="bwr", vmin=-1, vmax=1)
plt.xticks(range(4), pinguinos.columns, rotation=90)
plt.yticks(range(4), pinguinos.columns)
for i in range(len(pinguinos.columns)):
    for j in range(len(pinguinos.columns)):
        plt.text(i, j, round(matriz.iloc[i, j],2))
plt.colorbar()
plt.show()
```
