```python
import pandas as pd
```

```python
encabezado = ["tiempo", "volumen", "Género"]
musica = {"n1": [10, 45, "R"],
          "n2": [45, 84, "P"],
          "n3": [58, 14, "T"],
          "n4": [89, 24, "P"],
          "n5": [92, 94, "R"],
          "n6": [15, 4, "R"]}
musica = pd.DataFrame(musica, index=encabezado)
musica = musica.T
musica
```

```python
r = {"TiempoR": 0, "VolR": 0}
p = {"TiempoP": 0, "VolP": 0}
t = {"TiempoT": 0, "VolT": 0}
for i in range(len(musica)):
  if musica["Género"].iloc[i] == "R":
    r["TiempoR"] = r["TiempoR"] + musica["tiempo"].iloc[i]
    r["VolR"] = r["VolR"] + musica["volumen"].iloc[i]
  elif musica["Género"].iloc[i] == "P":
    p["TiempoP"] = p["TiempoP"] + musica["tiempo"].iloc[i]
    p["VolP"] = p["VolP"] + musica["volumen"].iloc[i]
  elif musica["Género"].iloc[i] == "T":
    t["TiempoT"] = t["TiempoT"] + musica["tiempo"].iloc[i]
    t["VolT"] = t["VolT"] + musica["volumen"].iloc[i]
# Porcentajes
pr = 100*r["TiempoR"]/(r["TiempoR"]+p["TiempoP"]+t["TiempoT"])
pp = 100*p["TiempoP"]/(r["TiempoR"]+p["TiempoP"]+t["TiempoT"])
pt = 100*t["TiempoT"]/(r["TiempoR"]+p["TiempoP"]+t["TiempoT"])
print("tus gustos son:",  int(pr), "%", "Rock")
print("tus gustos son:",  int(pt), "%", "Tango")
print("tus gustos son:",  int(pp), "%", "Pop")
```
